## 1. The RL Problem

Reinforcement Learning is about learning to make decisions:
- **Agent**: The learner (our chess player)
- **Environment**: The world (chess game)
- **State**: Current situation (board position)
- **Action**: What to do (which move)
- **Reward**: Feedback (win/loss/draw)
