# Chess Exploration with python-chess

This notebook helps you explore the python-chess library and understand how chess positions are represented.


```python
import chess
import chess.svg
from IPython.display import SVG, display
```

## Creating a Board


```python
# Create a new board in starting position
board = chess.Board()
print(board)
```

## Exploring Legal Moves


```python
# Get all legal moves
legal_moves = list(board.legal_moves)
print(f"Number of legal moves: {len(legal_moves)}")
print(f"First 10 moves: {[str(move) for move in legal_moves[:10]]}")
```

## Making Moves


```python
# Make a move
move = chess.Move.from_uci("e2e4")
board.push(move)
print("After e2e4:")
print(board)
```
